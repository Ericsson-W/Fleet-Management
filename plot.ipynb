### Plotting

```python
import pandas as pd 
import matplotlib.pyplot as plt

coord = pd.read_excel("vehicle_info.xlsx",sheet_name=1)
obstacles = pd.read_excel("vehicle_info.xlsx",sheet_name=2)
```

```python
plt.plot(obstacles['Location X'],-obstacles['Location Y'], 'ro', label='Obstacle')
plt.plot(coord['Target Location X'], coord['Target Location Y'], 'go', label='target path')
plt.plot(coord['Current Location X'], coord['Current Location Y'], 'b-', label='Vehicle Path')

plt.title(f'Path of Vehicle')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,3)
plt.ylim(-3,0) 
plt.grid(True)
plt.legend()
plt.show()
```
